```python
#Gender distribution
```

```python
plt.figure(figsize =(4,3))
sns.countplot(data= df,x="Gender")
plt.show()
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
sns.heatmap(gp, annot = True)
plt.show()
```

```python
ak= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(ak)
```

```python
sns.heatmap(ak, annot = True)
plt.title("relationshiop between parent ParentMaritalStatus and student score")
plt.show()
```

```python
groupA= df.loc[(df['EthnicGroup'] == "group A")].count()
print(groupA)
```
